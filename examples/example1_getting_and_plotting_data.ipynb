```python
import numpy as np
import matplotlib.pyplot as plt
import labrad_data_tools as ldt
```

# One dataset

```python
date = '20240718'
scantime = '1333.41'
exclude = []
```

```python
(x, y) = ldt.get_data(date, scantime, exclude=exclude)
expt_name = ldt.get_experiment_name(date, scantime)
```

```python
(fig, ax) = plt.subplots()
ax.plot(x, y, '.-', label=scantime)
ax.set_title('{} - {}'.format(date, expt_name))
ax.legend()
plt.show()
```

# Multiple datasets

```python
date = '20240718'
scantimes = ['1333.41', '1426.43']
excludes = [[], [-1, -2, -3]]
```

```python
expt_name = ldt.get_experiment_name(date, scantimes[0])
```

```python
(fig, ax) = plt.subplots()
for (scantime, exclude) in zip(scantimes, excludes):
    (x, y) = ldt.get_data(date, scantime, exclude=exclude)
    ax.plot(x, y, '.-', label=scantime)
ax.set_title('{} - {}'.format(date, expt_name))
ax.legend()
plt.show()
```
